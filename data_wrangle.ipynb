## Check for entries with more than one industry


```python
import pandas as pd
import ast

def safe_eval(x):
    if pd.isna(x):
        return []
    try:
        return ast.literal_eval(x)
    except:
        return []

def get_first_industry(industries_str):
    industries_list = safe_eval(industries_str)
    return industries_list[0].strip() if industries_list else None

# Find mismatches where both are present but different
mismatches = billionaire_df[
    (billionaire_df['business_category'].notna()) & 
    (billionaire_df['business_industries'].notna()) &
    (billionaire_df['business_category'] != billionaire_df['business_industries'].apply(get_first_industry))
]

print(f"Mismatches where both are present but different: {len(mismatches)}")

# Check if stripping whitespace resolves the discrepancies
resolved_by_stripping = mismatches[
    mismatches['business_category'].str.strip() == 
    mismatches['business_industries'].apply(lambda x: get_first_industry(x))
]

print(f"Discrepancies resolved by stripping whitespace: {len(resolved_by_stripping)}")

# Display some examples of resolved discrepancies
print("\nExamples of discrepancies resolved by stripping whitespace:")
print(resolved_by_stripping[['business_category', 'business_industries']].head())

# Check remaining discrepancies
remaining_discrepancies = len(mismatches) - len(resolved_by_stripping)
print(f"\nRemaining discrepancies after stripping whitespace: {remaining_discrepancies}")

if remaining_discrepancies > 0:
    print("\nExamples of remaining discrepancies:")
    remaining = mismatches[
        mismatches['business_category'].str.strip() != 
        mismatches['business_industries'].apply(lambda x: get_first_industry(x))
    ]
    print(remaining[['business_category', 'business_industries']].head())

# Check for case sensitivity issues in remaining discrepancies
case_insensitive_match = remaining[
    remaining['business_category'].str.strip().str.lower() == 
    remaining['business_industries'].apply(lambda x: get_first_industry(x).lower())
]

print(f"\nDiscrepancies resolved by case-insensitive comparison: {len(case_insensitive_match)}")

if len(case_insensitive_match) > 0:
    print("\nExamples of case-sensitivity discrepancies:")
    print(case_insensitive_match[['business_category', 'business_industries']].head())

# Final count of unexplained discrepancies
unexplained = remaining_discrepancies - len(case_insensitive_match)
print(f"\nUnexplained discrepancies: {unexplained}")

if unexplained > 0:
    print("\nExamples of unexplained discrepancies:")
    unexplained_df = remaining[
        remaining['business_category'].str.strip().str.lower() != 
        remaining['business_industries'].apply(lambda x: get_first_industry(x).lower())
    ]
    print(unexplained_df[['business_category', 'business_industries']].head())
```

### Drop the business_category column and replace business_industries with non-list version of itself

```python
import pandas as pd
import ast

def safe_eval_and_get_first(x):
    if pd.isna(x):
        return None
    try:
        industries = ast.literal_eval(x)
        return industries[0].strip() if industries else None
    except:
        return x  # Return the original string if it's not a valid list

# Apply the function to create a new column
billionaire_df['industry'] = billionaire_df['business_industries'].apply(safe_eval_and_get_first)

# Drop the original columns
billionaire_df = billionaire_df.drop(columns=['business_category', 'business_industries'])

# Display the first few rows to verify
print(billionaire_df[['last_name', 'first_name', 'industry']].head())

# Check for any remaining list-like strings
list_like = billionaire_df['industry'].str.contains(r'^\[.*\]$', na=False)
print("\nRows where 'industry' still looks like a list:")
print(billionaire_df[list_like][['last_name', 'first_name', 'industry']])

# Basic info about the new column
print("\nInformation about the new 'industry' column:")
print(billionaire_df['industry'].info())
print("\nTop 10 most common industries:")
print(billionaire_df['industry'].value_counts().head(10))
```

## Add country codes

```python
import pandas as pd
import pycountry
from functools import lru_cache

@lru_cache(maxsize=None)
def country_to_iso3(country_name):
    try:
        return pycountry.countries.search_fuzzy(country_name)[0].alpha_3
    except:
        return None

custom_mapping = {
    'Turkey': 'TUR',
    'St. Kitts and Nevis': 'KNA',
    'Swaziland': 'SWZ',
    'Macau': 'MAC',
    'Eswatini (Swaziland)': 'SWZ',
    'Hong Kong': 'HKG',
    'Hong Kong SAR, China': 'HKG',
    'Egypt, Arab Rep.': 'EGY',
    'Venezuela, RB': 'VEN',
    'Korea, Rep.': 'KOR'
}

def country_to_iso3_with_custom(country_name):
    if pd.isna(country_name) or str(country_name).strip() == '':
        return None
    if country_name in custom_mapping:
        return custom_mapping[country_name]
    return country_to_iso3(country_name)

# Read the CSV file
billionaire_df = pd.read_csv("billionaire_data.csv")

# Identify rows with NaN in country_of_citizenship
nan_citizenship = billionaire_df[billionaire_df['country_of_citizenship'].isna()]

# Function to find country for a person in adjacent years
def find_country_in_adjacent_years(full_name, year):
    for adj_year in range(year - 2, year + 3):  # Check 2 years before and after
        country = billionaire_df[(billionaire_df['full_name'] == full_name) & 
                                 (billionaire_df['year'] == adj_year) & 
                                 (billionaire_df['country_of_citizenship'].notna())]['country_of_citizenship'].values
        if len(country) > 0:
            return country[0], adj_year
    return None, None

# Check for countries in adjacent years
print("Checking for missing citizenships in adjacent years:")
for index, row in nan_citizenship.iterrows():
    full_name = row['full_name']
    year = row['year']
    country, found_year = find_country_in_adjacent_years(full_name, year)
    if country:
        print(f"  {full_name} ({year}): Found citizenship '{country}' in year {found_year}")
        billionaire_df.at[index, 'country_of_citizenship'] = country
    else:
        print(f"  {full_name} ({year}): No citizenship found in adjacent years")

# Apply ISO3C conversion
billionaire_df['iso3c'] = billionaire_df['country_of_citizenship'].apply(country_to_iso3_with_custom)

# Final check for remaining NAs
remaining_na = billionaire_df[billionaire_df['country_of_citizenship'].isna()]
print(f"\nRemaining NAs after checking adjacent years: {len(remaining_na)}")

if len(remaining_na) > 0:
    print("Sample of remaining NAs:")
    print(remaining_na[['full_name', 'year', 'country_of_citizenship', 'iso3c']].head())

# Save the updated DataFrame to a new CSV file
billionaire_df.to_csv('billionaire_data_updated.csv', index=False)
print("\nUpdated data saved to 'billionaire_data_updated.csv'")
```

Relocate the iso3c column to the right after the country_of_citizenship column.

```python
def relocate(df, column, after):
    """
    Relocate a column in a DataFrame.
    
    :param df: pandas DataFrame
    :param column: str, name of the column to move
    :param after: str, name of the column to move the target column after
    :return: pandas DataFrame with relocated column
    """
    col = df.pop(column)
    df.insert(df.columns.get_loc(after) + 1, column, col)
    return df

# Use the function to move 'iso3c' after 'country_of_citizenship'
billionaire_df = relocate(billionaire_df, 'iso3c', 'country_of_citizenship')

# Verify the new column order
print("\nNew column order:")
print(billionaire_df.columns.tolist())

# Save the updated DataFrame to a new CSV file
billionaire_df.to_csv('billionaire_data_updated.csv', index=False)
print("\nUpdated data saved to 'billionaire_data_updated.csv'")
```

## Get WB data

```python
import wbdata
import pandas as pd
from datetime import datetime

# Define the indicators
indicators = {
    'NY.GDP.MKTP.CD':  'gdp_pc_current_usd',
    'NY.GDP.MKTP.KD':  'gdp_pc_constant_2015_usd',
    'NY.GDP.PCAP.PP.KD': 'gdp_per_capita_ppp_constant',
    'NY.GDP.DEFL.ZS.AD': 'gdp_deflator',  
    'PA.NUS.FCRF': 'official_exchange_rate',
    'SP.POP.TOTL': 'population'  # Added for filtering
}

# Set the date range
current_year = datetime.now().year
date_range = (datetime(1997, 1, 1), datetime(current_year, 1, 1))

# Fetch World Bank data for all countries from 1997 to the most recent year
wb_data = wbdata.get_dataframe(indicators, country=['all'], date=date_range)

# Reset index to make 'country' and 'date' columns
wb_data = wb_data.reset_index()

# Identify aggregates
geos = pd.Series(wb_data.country.unique())
world_index = geos[geos == 'World'].index[0]
aggs = geos[:world_index+1]

# Filter out aggregates
wb_data = wb_data[~wb_data.country.isin(aggs)]

# Rename columns
wb_data = wb_data.rename(columns={'country': 'country_name', 'date': 'year'})

# Convert year to datetime and then to integer
wb_data['year'] = pd.to_datetime(wb_data['year']).dt.year

# Add ISO3C codes using the existing function
wb_data['iso3c'] = wb_data['country_name'].apply(country_to_iso3_with_custom)

# Use the relocate function to move 'iso3c' after 'country_name'
wb_data = relocate(wb_data, 'iso3c', 'country_name')

# Display the first few rows and basic info
print(wb_data.head())
print("\nDataset Info:")
print(wb_data.info())

# Display the range of years in the dataset
print(f"\nYear range in the dataset: {wb_data['year'].min()} to {wb_data['year'].max()}")

# Count of countries
print(f"\nNumber of countries: {wb_data['country_name'].nunique()}")

# Check for countries without ISO3C codes
missing_codes = wb_data[wb_data['iso3c'].isnull()]
if not missing_codes.empty:
    print("\nCountries without ISO3C codes:")
    print(missing_codes['country_name'].unique())


```

## Join the data frames

```python
# Perform the merge with an indicator
merged_df = billionaire_df.merge(wb_data, on=['iso3c', 'year'], how='left', indicator=True)

# Identify unmatched rows
unmatched = merged_df[merged_df['_merge'] == 'left_only']

if not unmatched.empty:
    print("\nTotal number of unmatched rows:", len(unmatched))
    
    # Count unmatched cases by country code
    unmatched_counts = unmatched['iso3c'].value_counts(dropna=False)
    
    # Count unmatched cases from 2024 by country code
    unmatched_2024 = unmatched[unmatched['year'] == 2024]['iso3c'].value_counts(dropna=False)
    
    # Calculate percentage of 2024 cases
    percent_2024 = (unmatched_2024 / unmatched_counts * 100).round(2)
    
    # Combine counts and percentages
    unmatched_summary = pd.DataFrame({
        'Total Unmatched': unmatched_counts,
        'Percent from 2024': percent_2024
    }).fillna(0)  # Fill NaN with 0 for countries with no 2024 cases
    
    # Separate NA counts
    na_count = unmatched_counts.get(pd.NA, 0)  # Use 0 if no NAs
    
    print("\nNumber of unmatched cases with missing country code (NA):", na_count)
    
    print("\nNumber of unmatched cases by country code and percent from 2024:")
    # Set display options to show all rows
    with pd.option_context('display.max_rows', None):
        print(unmatched_summary.sort_values('Total Unmatched', ascending=False))
    
    print("\nSample of all unmatched rows:")
    print(unmatched[['full_name', 'year', 'iso3c']].head())
else:
    print("\nAll rows were successfully matched.")
```
